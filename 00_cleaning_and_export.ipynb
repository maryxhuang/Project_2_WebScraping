```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('reviews.csv', encoding = 'iso-8859-1')
df.head()
```

```python
df.dtypes
```

```python
df.iloc[2, 7], df.iloc[2, 8], df.iloc[3, 7], df.iloc[3, 7]
```

```python
df['Distance'] = df['Distance'].apply(lambda x: re.sub('[\x97]', '', x))
df['Distance'] = df['Distance'].apply(lambda x: re.sub('[\x97]', '', x))
df['Cost'] = df['Cost'].apply(lambda x: re.sub('[\x97]', '', x))
df['Per.Mile'] = df['Per.Mile'].apply(lambda x: re.sub('[\x97]', '', x))
df['Sqft.Moved'] = df['Sqft.Moved'].apply(lambda x: re.sub('[\x97]', '', x))
df.head()
```

```python
df.iloc[2, 7], df.iloc[2, 8], df.iloc[3, 7], df.iloc[3, 7]
```

```python
np.sum(df.isnull())
```

```python
np.sum(df.isnull())
```

```python
df.head()
```

```python
df.rename(columns={'Per.Mile': 'Cost.per.Mile'}, inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df['Distance'] = df['Distance'].apply(lambda x: float(''.join(x.split()[0].split(','))) if x != '' else np.nan)
df['Cost'] = df['Cost'].apply(lambda x: float(''.join(x.strip('$').split(','))) if x != '' else np.nan)
df['Cost.per.Mile'] = df['Cost.per.Mile'].apply(lambda x: float(''.join(x.strip('$').split(','))) if x != '' else np.nan)
df['Sqft.Moved'] = df['Sqft.Moved'].apply(lambda x: float(''.join(x.split(','))) if x != '' else np.nan)
df.head()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('reviews_cleaned.csv')
df.head()
```

```python
df_milecost = df[['Company', 'Cost.per.Mile', 'Sqft.Moved', 'Distance', 'Cost']]
df_milecost.shape
```

```python
df_milecost = df_milecost.dropna()
df_milecost.shape
```

```python
g = sns.lmplot('Distance', 'Cost', df_milecost, col = 'Company', 
           hue = 'Company', palette = 'Set1', size = 3, aspect = 0.5)
```

```python
g = sns.lmplot('Distance', 'Cost.per.Mile', df_milecost, col = 'Company', fit_reg=False,
           hue = 'Company', palette = 'Set2', size = 3, aspect = 1)
```

```python
g = sns.lmplot('Distance', 'Cost', df_milecost, col = 'Company', 
           hue = 'Company', palette = 'Set1', size = 6, aspect = 0.5, sharex=False, sharey=False)
g.set(xlim=(-50, 4000), ylim=(-50, 5500))
```

```python
g = sns.lmplot('Distance', 'Cost.per.Mile', df_milecost, col = 'Company', fit_reg=False,
           hue = 'Company', palette = 'Set2', size = 3, aspect = 1, sharex=False, sharey=False)
g.set(xlim=(-50, 4000), ylim=(-2, 30))
```

```python
df_milecost3 = df[['Company', 'Distance', 'Cost', 'Cost.per.Mile']]
df_milecost3 = df_milecost3[(df_milecost3['Company'] != 'Enterprise') & (df_milecost3['Company'] != 'Ryder')]
df_milecost3 = df_milecost3[df_milecost3['Distance'].notnull() & df_milecost3['Cost'].notnull() & df_milecost3['Cost.per.Mile'].notnull()]
df_milecost3.shape
```

```python
df_milecost3.sample(5)
```

```python
df_milecost3_loc = df_milecost3[df_milecost3['Distance'] <= 50]
df_milecost3_loc.shape
```

```python
g = sns.lmplot('Distance', 'Cost', df_milecost3_loc, markers="x", scatter_kws={"s":10, 'alpha':0.3},
           hue = 'Company', palette = 'Set1', size = 6, aspect = 1, sharex=False, sharey=False)
g.set(xlim=(0, 55), ylim=(0, 500))
```

```python
g = sns.lmplot('Distance', 'Cost.per.Mile', df_milecost3_loc, markers="x", scatter_kws={"s":10, 'alpha':0.3}, fit_reg=False,
           hue = 'Company', palette = 'Set2', size = 6, aspect = 1, sharex=False, sharey=False)
g.set(xlim=(-2, 55), ylim=(-10, 300))
```

```python
df_milecost3_med = df_milecost3[(df_milecost3['Distance'] > 50) & (df_milecost3['Distance'] <= 250)]
df_milecost3_med.shape
```

```python
g = sns.lmplot('Distance', 'Cost', df_milecost3_med, markers="x", scatter_kws={"s":10, 'alpha':0.3}, 
           hue = 'Company', palette = 'Set1', size = 6, aspect = 1, sharex=False, sharey=False)
g.set(xlim=(45, 255), ylim=(0, 1000))
```

```python
g = sns.lmplot('Distance', 'Cost.per.Mile', df_milecost3_med, markers="x", scatter_kws={"s":10, 'alpha':0.3}, fit_reg=False,
           hue = 'Company', palette = 'Set2', size = 6, aspect = 1, sharex=False, sharey=False)
g.set(xlim=(45, 255), ylim=(-1, 10))
```

```python
df_milecost3_long = df_milecost3[df_milecost3['Distance'] > 250]
df_milecost3_long.shape
```

```python
g = sns.lmplot('Distance', 'Cost', df_milecost3_long, markers="x", scatter_kws={"s":10, 'alpha':0.3},
           hue = 'Company', palette = 'Set1', size = 6, aspect = 1, sharex=False, sharey=False)
g.set(xlim=(240, None), ylim=(0, 5000))
```

```python
g = sns.lmplot('Distance', 'Cost.per.Mile', df_milecost3_long, markers="x", scatter_kws={"s":10, 'alpha':0.3}, fit_reg=False,
           hue = 'Company', palette = 'Set2', size = 6, aspect = 1, sharex=False, sharey=False)
g.set(xlim=(250, None), ylim=(-1, 10))
```

```python
current_palette = sns.color_palette("RdBu_r", 15)
sns.palplot(current_palette)
```

```python
sns.set(rc={"figure.figsize": (5, 5)})
sns.heatmap(df_num.corr(), cmap = current_palette, vmax=1, vmin=-1, square=True).set_title('Moving Review Correlations')
```
